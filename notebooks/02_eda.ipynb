```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
pd.set_option("display.max_columns", None)

df = pd.read_csv("cleaned_used_cars.csv")
df.shape

plt.figure(figsize=(8,5))
sns.boxplot(x="vehicle_age", y="selling_price", data=df)
plt.title("Vehicle Age vs Selling Price")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x="km_driven", y="selling_price", data=df, alpha=0.4)
plt.title("Kilometers Driven vs Selling Price")
plt.show()

top_brands = df["brand"].value_counts().head(10).index

plt.figure(figsize=(10,5))
sns.boxplot(
    data=df[df["brand"].isin(top_brands)],
    x="brand",
    y="selling_price"
)
plt.xticks(rotation=45)
plt.title("Top Brands vs Selling Price")
plt.show()

plt.figure(figsize=(7,4))
sns.boxplot(x="fuel_type", y="selling_price", data=df)
plt.title("Fuel Type vs Price")
plt.show()

plt.figure(figsize=(7,4))
sns.boxplot(x="transmission_type", y="selling_price", data=df)
plt.title("Transmission Type vs Price")
plt.show()

plt.figure(figsize=(7,4))
sns.boxplot(x="seller_type", y="selling_price", data=df)
plt.title("Seller Type vs Price")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x="max_power", y="selling_price", data=df, alpha=0.4)
plt.title("Engine Power vs Price")
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=False,
    cmap="coolwarm"
)
plt.title("Feature Correlation Heatmap")
plt.show()
```
